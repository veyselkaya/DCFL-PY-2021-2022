```python
print("Mükemmel sayıyı bulan python programı...")
print(50*"*")


sayi = int(input("Kontrol etmek istediğiniz sayıyı giriniz: "))
i=1
toplam =0
liste = []

while i<sayi:
  if sayi %i==0:
    toplam +=i
    liste.append(i)
  i +=1

if toplam == sayi:

  print("Girdiğin sayı:", sayi, "bir mükemmel sayıdır")
  print("Girdiğiniz sayının kendisi hariç çarpanları. ", liste)

else:
  print(f"Girdiğin sayı: {sayi} bir mükemmel sayı değildir")
```

```python
def faktoriyel(sayi):
  faktoriyel =1
  print(f"Girdiğin sayı {sayi}")
  if sayi == 0 or sayi ==1:
    print(f"Girdiğin sayı {sayi} faktöriyeli: ", faktoriyel)
  else:
    while sayi >=1:
      faktoriyel *=sayi
      sayi -=1
    print(f"Girdiğin sayının faktöriyeli: ", faktoriyel)
    print("Sayısı: ", len(str(faktoriyel)), "basamaklıdır")
 
```

```python
faktoriyel(5)
```

```python
faktoriyel(8)
```

```python
faktoriyel(45651)
```
